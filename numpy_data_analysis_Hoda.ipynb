### A list of total cases and total deaths of each country


```python
new_regions = np.unique(data[:, 5]) #Countries after dropping missing values
country_total_mat = []
for country in new_regions:
    filtered_data = data[np.where(data[:, 5] == country)]
    total_cases = np.sum(filtered_data[:, 3])
    total_deaths = np.sum(filtered_data[:, 4])
    country_total_mat.append([total_cases, total_deaths, np.unique(filtered_data[:, 8])[0], country])
country_total_mat = np.array(country_total_mat)
country_total_mat
    
```

### The relationship between the number of total cases, number of total deaths and the population
By looking at the results, it can be concluded that the number of deaths is highly related to the number of cases, but the number of cases is not so related to the population of the countries, so is the number of deaths to the population.

```python
np.corrcoef(country_total_mat[:, :3].astype(float), rowvar=False)
```

### mean median std for each country
We can see that for many countries, the median is much smaller than the mean, which means that the growth in the number of cases and deaths has not been steady through the days and has been concentrated in the last days. The high standard deviations means that the distribution of the numbers is not concentrated around the mean, so the number of cases in different days are very diverse and not very close to the average. 

```python
mean_mat = []
for country in new_regions:
    filtered_data = data[np.where(data[:, 5] == country)]
    meanof_cases = np.mean(filtered_data[:, 3])
    meanof_deaths = np.mean(filtered_data[:, 4])
    mean_mat.append([math.ceil(meanof_cases), math.ceil(meanof_deaths), country])
mean_mat = np.array(mean_mat)
mean_mat
```

```python
median_mat = []
for country in new_regions:
    filtered_data = data[np.where(data[:, 5] == country)]
    medof_cases = np.median(filtered_data[:, 3])
    medof_deaths = np.median(filtered_data[:, 4])
    median_mat.append([medof_cases, medof_deaths, country])
median_mat = np.array(median_mat)
median_mat
```

```python
std_mat = []
for country in new_regions:
    filtered_data = data[np.where(data[:, 5] == country)]
    stdof_cases = np.std(filtered_data[:, 3])
    stdof_deaths = np.std(filtered_data[:, 4])
    std_mat.append([round(stdof_cases), round(stdof_deaths), country])
std_mat = np.array(std_mat)
std_mat
```
